```python
"""
Cordell Newmiller
Prof. Fassnacht
PHY 266
Homework 2
Problem 2

In this problem, you are going to see how quickly a Poisson distribution converges to a Gaussian,
especially in the wings of the distribution. Use python’s random number generator to create five
data sets with 10,000 members each. The data in the five samples should be drawn from Poisson
distributions with means of 3, 10, 30, 100, and 1000, respectively. As usual, submit all your code
for this problem.
"""

import numpy

# When debugging, it's useful to decrease the number of data points used N. 
# The problem calls for N=10000.
N=10000
Means = numpy.array([3, 10, 30, 100, 1000])

# This creates an array with N rows, and one column for each of the desired means. 
RandomDataSet =  numpy.random.poisson(Means,size=(N,Means.size))


"""
Part (a): For each of the data sets, calculate the percentage of points that are contained within the
µ ± nσ regions, where n = 1, 2, 3. Make a table, where the rows are “1 σ”, “2 σ”, and “3 σ” and
the columns are the [five] data sets.
"""

# This problem calls for analysis within 1, 2, and 3 sigmas
WithinHowManySigmas=3

# For Poisson distributions, the mean is equal to the varience
Sigmas=Means**(1/2)

# This table will record the number of values falling within the intervals
NumberWithinTable=numpy.zeros([WithinHowManySigmas,Means.size])


# For each row i, add up the number of elements that fall within the intervals
for i in range(1,WithinHowManySigmas+1):
    NumberWithinTable[i-1,:]=(numpy.sum(numpy.logical_and((
                    Means-i*Sigmas)<RandomDataSet, RandomDataSet<(Means+i*Sigmas)),axis=0))
    # We have to use Numpy's logical_and to compare arrays.
    # The comparison appears to me to be a dimension mismatch at first, 
    # but Python recognizes the intended operation and compares each row. 


PercentWithinTable=NumberWithinTable/N
print('Part (a), table of percentages: \n',PercentWithinTable)

"""
Part (b): Find the boundaries of the regions that contain 68%, 95.4%, and 99.7% of the data sets that
you generated. Compare these boundaries to those defined by the corresponding µ ± nσ regions.
"""
print('Part (b), boundaries')

# Sort each column
SortedDataSet=numpy.msort(RandomDataSet)

Percentages=numpy.array([.68,.954,.997])

# Initialize tables
LowerBounds=numpy.zeros([WithinHowManySigmas,Means.size])
UpperBounds=numpy.zeros([WithinHowManySigmas,Means.size])
TheoreticalLowerBounds=numpy.zeros([WithinHowManySigmas,Means.size])
TheoreticalUpperBounds=numpy.zeros([WithinHowManySigmas,Means.size])

# Fill tables
for i in range(1,WithinHowManySigmas+1):
    LowerBounds[i-1,:]= SortedDataSet[int(numpy.around(N*(1-Percentages[i-1]))/2),:]
    UpperBounds[i-1,:]= SortedDataSet[int(numpy.around(N*(1+Percentages[i-1]))/2),:]
    TheoreticalLowerBounds[i-1,:]=Means-i*Sigmas
    TheoreticalUpperBounds[i-1,:]=Means+i*Sigmas

# Clip out negative theoretical bounds
TheoreticalLowerBounds=TheoreticalLowerBounds.clip(min=0)

numpy.set_printoptions(suppress=True,precision=3) # Suppress scientific notation output
print('Actual lower bounds: \n',LowerBounds)
print('Theoretical lower bounds: \n',TheoreticalLowerBounds)
print('Actual upper bounds: \n',UpperBounds)
print('Theoretical upper bounds: \n',TheoreticalUpperBounds)

print('Percent differences of lower bounds \n',
      (LowerBounds.clip(min=.001)/TheoreticalLowerBounds.clip(min=.001) - 1)*100) 
      # Avoid dividing by 0
print('Percent differences of upper bounds \n',(UpperBounds/TheoreticalUpperBounds - 1)*100)

numpy.set_printoptions(suppress=False,precision=8) # Reset to default, useful when working in notebook

# End of problem 2
```
