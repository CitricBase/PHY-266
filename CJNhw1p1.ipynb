```python
"""
Cordell Newmiller
PHY 266
Prof. Fassnacht
Homework 1
Problem 1

(a) Calculate the mean and standard deviation of the data in each of the 3 files.
(b) Each of the 3 files contains a 2048×2048 data array. Create a new 2048×2048 data array that 
is the mean of the three input data arrays. In other words, each pixel in the new array should 
have a data value that is the mean of the data values in that same pixel in each of the three 
input arrays. Calculate the mean and standard deviation of the data in this new array.
"""

# Set up environment
import numpy
from astropy.io import fits

# Load the data. This program assumes the fits files are in a directory named PHY266 in 
# the working directory.
bias1=fits.open('PHY266/bias1.fits')
bias2=fits.open('PHY266/bias2.fits')
bias3=fits.open('PHY266/bias3.fits')

# Pull out the data arrays
a1=bias1[0].data
a2=bias2[0].data
a3=bias3[0].data

"""
Part (a) of the problem! Calculate means and standard deviations. I'm assuming 
I'm  not supposed to just use the built-in mean and stdev functions to do this. 
"""

# Calculate the mean of an array
def mymean(x,check=False):
    x = numpy.zeros(x.shape) + x   # Turns x to floats to prevent int overflow
    m = x.sum()/x.size
    if check:     # check the result against what it's actually supposed to be, for debugging
        print('Sanity check... ', end="") # end trick to prevent newline
        if abs(m-x.mean())<0.00001:
            print('passed')
        else:
            print('failed')
            print(m, 'vs', x.mean())
    return m

# Deliver results
print('The mean of the first  dataset is', mymean(a1,True))
print('The mean of the second dataset is', mymean(a2))
print('The mean of the third  dataset is', mymean(a3))

# Calculate the standard deviation of an array
def mystdev(x,check=False):
    s = (numpy.sum((x-mymean(x)) ** 2)/(x.size-1)) ** (1/2)    # Python uses ** to mean power
    if check:     # check the result against what it's actually supposed to be
        print('Sanity check... ', end="")    # Note: numpy stdev doesn't subtract one...?
        if abs(s-x.std()) < 0.00001:
            print('passed')
        else:
            print('failed')
            print(s, 'vs', x.std())
    return s

# Deliver results
print('The standard deviation of the first  dataset is', mystdev(a1,True))
print('The standard deviation of the second dataset is', mystdev(a2))
print('The standard deviation of the third  dataset is', mystdev(a3))


"""
Part (b) of the problem. Average the three datasets together, and show statistics on the 
resulting combined set.
"""

# It might be faster and more general to use a 3D matrix and numpy's mean along one 
# dimension, but my mymean function doesn't do that trick.
aaa= (a1 + a2 + a3)/3

print('The mean of the means is', mymean(aaa))
print('The standard deviation of the means is', mystdev(aaa))

# End of problem 1.
```
