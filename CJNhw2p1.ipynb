```python
"""
Cordell Newmiller
Prof. Fassnacht
PHY 266
Homework 2
Problem 1

You have a box containing Ntot = 9 wooden blocks, NB of which are blue and NR of which are
red. You will do the classic thing, which is to randomly draw one block at a time, each time
replacing the block that you have drawn back into the box before picking the next block.
"""

"""
Part (a): On any given draw, what is the probability of getting a red block? Note that this won’t be
a number, but rather an expression.
"""

print('Part (a): the probability of drawing a red block from a pool of 9',
      'blocks, \n Nr of which are red, is p = Nr/9.\n')

"""
Part (b): If you draw M times, with replacement, what is the probability of getting R red blocks?
Here R obviously has to have a value between 0 and M. (This is the classical statistics question)
"""

print('Part (b): this is a binomial probability. \n P = (M choose R) * p^R * (1-p)^(M-R)\n')

"""
Part (c): Remember that the likelihood is defined as P(data|model). If you draw six times and get
two red blocks, what value of NR maximizes the likelihood?
"""

print('Part (c): the most likely value of Nr is the one that matches the model probability with \n', 
      'the data\'s probability. The data\'s probability is 2/6=1/3, so the most likely value of Nr \n',
      'would be Nr/9=1/3 --> Nr=3. \n',
      '\n You can also do this by differentiating the answer to part (b) with respect to Nr and \n', 
      'setting it equal to 0, with M=6, R=2, and p=Nr/9, you get the same result of Nr=3.')

"""
Part (d) 

Now we get to be Bayesians. You draw six blocks, of which two are red. Do the following
two things: (i) Calculate and report the values of the posterior probability, i.e., P(NR|data) for
each value NR = 0, 1, 2, . . . 9. Here “data” refers to the fact that you got 2 red balls out of 6
draws. Assume a uniform prior on NR, i.e., that NR is equally likely to have any of the values
between 0 and 9. Note that this posterior needs to be normalized such that
sum(P(i|data)) = 1
(ii) Plot the results, with NR along the x-axis and P(NR|data) on the y-axis.
"""

import numpy
from matplotlib import pyplot
from scipy.special import comb  # Combinatorial function
get_ipython().magic('matplotlib inline')


# Bayes' Theorem: P(Ai|B) = P(B|Ai)*P(Ai) / sum( P(B|Aj)*P(Aj) )
# In our case, Ai are the possible models, and B is the data.

def posteriors(M,R,N=9):
    Nr=numpy.arange(0,N+1)
    p=Nr/N
    PA=1/(Nr.size)                          # each model is equally likely
    PBA= comb(M,R) * (p**R) * (1-p)**(M-R)  # vector
    # Use Bayes' theorem to calculate the normalized posterior probability
    # even though, in this case, it turns out that everything cancels and PAB = PBA
    # because each model is equally likely to begin with
    PAB= PBA * PA / (numpy.sum(PBA * PA))
    
    pyplot.plot(Nr,PAB,'bo-')
    pyplot.xlabel('Number of actual red blocks')
    pyplot.ylabel('Probability')
    return

figure1 = pyplot.figure(figsize=(10,10))
pyplot.title('Part (d), drawing two red blocks out of six draws')
posteriors(6,2)


"""
(e) Do the same thing as in (d), but now you have gotten 20 red blocks in 60 draws.

"""
figure2 = pyplot.figure(figsize=(10,10))
pyplot.title('Part (e), drawing twenty red blocks out of sixty draws')
posteriors(60,20)

# End of problem 1
```
